```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.colors import LogNorm
```

```python
EX_DIR = '../extract_cr/extracted/'
```

```python
ls ../extract_cr/extracted/10/300/
```

```python
def get_ave(year, night, side, div):
    path = os.path.join(EX_DIR, '{year}/{night}/ave_divs_all_{year:02d}_{night:03d}_{side}_{div:02d}.fits')
    path = path.format(year=year, night=night, side=side, div=div)
    data = fits.getdata(path)
    return data
```

```python
plt.imshow(get_ave(10, 320, 'left', 31), norm=LogNorm(), cmap='magma')
plt.colorbar()
```

```python
data = get_ave(10, 320, 'left', 0)
plt.imshow(data[11:]-data[:10][::-1] - np.min(data[11:]-data[:10][::-1]) + 1, norm=LogNorm())
plt.colorbar()
```

```python
for i in range(32):
    data = get_ave(10, 320, 'right', i)
    plt.plot(data[11:, 10] - data[:10, 10][::-1], color=plt.cm.viridis(i*8))
```
